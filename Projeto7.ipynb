### Limpeza e transformação dos dados

```python
# Verificando valores nulos (Ausentes) no DataFrame
df.isnull().sum()
```

```python
# Checando as colunas que tem valor = '?'
df.isin(['?']).any()
```

```python
# Verificando valores únicos por coluna
for col in list(df.columns):
    # Obter uma lista de valores únicos
    list_unique_values = df[col].unique()
    # Se o número de valores exclusivos for menor que 15, imprima os valores.
    # Caso contrário, imprima o número de valores exclusivos
    if len(list_unique_values) < 15:
        print("\n")
        print(col + ': ' + str(len(list_unique_values)) + ' valores únicos')
        print(list_unique_values)
    else:
        print("\n")
        print(col + ': ' + str(len(list_unique_values)) + ' valores únicos')
```

Importante perceber que o valor ausente pode aparecer como "?".

```python
# Verificando a quantidade e percentual de registros onde o valor é igual a '?'
for col in df.columns:
    if df[col].dtype == object:
        if df[col][df[col] == '?'].count() > 0:
            print('\nColuna', col, 'tem', df[col][df[col] == '?'].count(), 'valores com o caracter "?"')
            print('Isso representa', round(df[col][df[col] == '?'].count() / len(df.index) * 100, 2), '% do total')
```

```python
# A coluna gender também apresenta um valor que precisa ser tratado
print('\nColuna gender tem', df['gender'][df['gender'] == 'Unknown/Invalid'].count(), 'valores com "Unknown/Invalid"')
print('Isso representa', round(df['gender'][df['gender'] == 'Unknown/Invalid'].count() / len(df.index) * 100, ))
```

**Agora a parte mais importante - decidir o que fazer!**

- Há valores ausentes sobre o peso do paciente em mais de 96% dos registro na variável 'weight'.
- O código do pagador e especificidade médica também apresentam valores ausentes de 40 a 50%.
- As demais variáveis apresentam um baixo percentual de valores ausentes.

  O que fazer?

**Alternativas:**

- Colocar a variável peso como categórica com dois labels: disponível/não disponível (1 e 0).
- Criar um código genérico para a variável payer_code como 99 e preencher os  valores ausentes.
- Criar um tipo genérico para a variável medical_specialty como 'Sem Especialidade Definida' e preencher os valores ausentes.
- Para a variável gender são apenas 3 registros e o ideal é removê-los.

**Decisões:**

- Devida a fraca interpretabilidade da variável peso, o melhor é simplesmente  ignorar a variável e removÊ-la.
- Descartar as variávis payer_code e medical_specialty, pois seria difícil tratar quantidade muito grande de valores ausentes.
- Remoção dos registro com '?' das demais variáveis.
- Para a variável gender vamos remover os 3 registros com categoria "Unknown/Invalid".

```python
# Removendo as 3 colunas com alto percentual de valores ausentes
df = df.drop(['weight', 'payer_code', 'medical_specialty'], axis=1)
```

**Para as demais variáveis com baixo percentual de valores ausentes, removeremos os registros**

```python
# Remoção dos registros com baixo percentual de valores ausentes
df = df[df['race'] != '?']
df = df[df['diag_1'] != '?']
df = df[df['diag_2'] != '?']
df = df[df['diag_3'] != '?']
df = df[df['gender'] != 'Unknown/Invalid']
```

```python
# Checando as colunas para verificar se há algum valor igual a '?'
df.isin(['?']).any()
```

**Vamos eliminar as colunas que têm apenas um único valor, pois elas não adicionam nada à análise. Queremos analisar variáveis e não constantes.**

```python
# Calculando o número total de valores em cada coluna
total_valores = df.shape[0]

# Calculando o número de valores ausentes em cada  coluna
valores_ausentes = df.isnull().sum()

# Calculando o percentual de valores ausentes em cada coluna
percentual_ausentes = (valores_ausentes / total_valores) * 100

# Exibe somente colunas com percentual maior do que zero
percentual_ausentes[percentual_ausentes > 0]
```
